```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies#Selected_changes_to_the_list_of_S&P_500_components'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
soup.find_all('table')
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
# <table class="wikitable sortable jquery-tablesorter" id="constituents">


table = soup.find('table', class_ = 'wikitable sortable')
```

```python
company_titles = table.find_all('th')
```

```python
company_titles
```

```python
comp_table_titles = [title.text.strip() for title in company_titles]

print(comp_table_titles)
```

```python
df = pd.DataFrame(columns = comp_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df.head(5)
```

```python
df.to_csv(r'D:\PortfolioProjects\WebScrapping\S&P 500\List_S&P500.csv', index = False)
```
